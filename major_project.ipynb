```python
from PIL import Image, ImageDraw, ImageFont

def create_fake_chat(excuse):
    img = Image.new('RGB', (500, 300), color=(255, 255, 255))
    draw = ImageDraw.Draw(img)
    draw.text((20, 100), "You: " + excuse, fill=(0, 0, 0))
    img.save("chat_proof.png")

create_fake_chat(excuse)
```

```python
import gradio as gr

prompt_map = {
    "work": "I can't come to work today because",
    "school": "I missed school today due to",
    "social": "I'm unable to attend the party because"
}

def generate_excuse(prompt):
    from transformers import pipeline
    generator = pipeline("text-generation", model="gpt2")
    output = generator(prompt, max_length=60, num_return_sequences=1)
    return output[0]['generated_text']

def excuse_interface(scenario):
    prompt = prompt_map[scenario]
    return generate_excuse(prompt)

dropdown_input = gr.Dropdown(choices=list(prompt_map.keys()), label="Select Scenario")

gr.Interface(
    fn=excuse_interface,
    inputs=dropdown_input,
    outputs="text",
    title="AI-Powered Excuse Generator"
).launch()
```
